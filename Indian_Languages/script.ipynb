```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests

header = {'User-Agent':"Mozilla/5.0 (X11; CrOS x86_64 8172.45.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.64 Safari/537.36"}
```

```python
URL1 = 'https://en.wikipedia.org/wiki/Languages_with_official_status_in_India#List_of_official_languages_by_states_and_territories'
# URL2 ='https://www.distancelatlong.com/country/india'
```

```python
r = requests.get(URL1, headers = header)
soup = BeautifulSoup(r.content,'html5lib')
```

```python
soup
```

```python
tables  = soup.find_all('table')
# tables = soup.find_all('table')
```

```python
# table = tables[2]
```

```python
by_states = tables[1]
by_territories = tables[2]
scripts = tables[3]
```

```python
tds = [row.text for row in by_states.findAll('td')]
# tds = [row.text for row in table.findAll('td')]
```

```python
tds
```

```python
states = [tds[i] for i in range(0,len(tds)) if i%4 == 1]
states
len(states)
# states = [tds[i] for i in range(0,len(tds)) if i%3 == 0]
# states
# len(states)
```

```python
Off_lang = [tds[i] for i in range(0,len(tds)) if i%4==2]
Off_lang
len(Off_lang)
# lattitudes = [tds[i] for i in range(0,len(tds)) if i%3==1]
# lattitudes
# len(lattitudes)
```

```python
Other_offs = [tds[i] for i in range(0,len(tds)) if i%4==3]
Other_offs
len(Other_offs)
# longitudes = [tds[i] for i in range(0,len(tds)) if i%3==2]
# longitudes
# len(longitudes)
```

```python
d = {'state':states, 'Official Language(s)':Off_lang, 'Other Official Language(s)':Other_offs}
states_df = pd.DataFrame(data=d)
# d1 = {'state':states,'lattitude':lattitudes,'longitude':longitudes}
```

```python
tds = [row.text for row in by_territories.findAll('td')]
```

```python
territories = [tds[i] for i in range(0,len(tds)) if i%4 == 1]
territories
len(territories)
```

```python
Off_lang = [tds[i] for i in range(0,len(tds)) if i%4==2]
Off_lang
len(Off_lang)
```

```python
Other_offs = [tds[i] for i in range(0,len(tds)) if i%4==3]
Other_offs
len(Other_offs)
```

```python
d= {'territory':territories, 'Official Language(s)':Off_lang, 'Other Official Language(s)':Other_offs}
territory_df = pd.DataFrame(data=d)
```

```python
tds = [row.text for row in scripts.findAll('td')]
```

```python
scripts = [tds[i] for i in range(0,len(tds)) if i%2==0 or i%2==2]
```

```python
languages = [tds[i] for i in range(0,len(tds)) if i%2==1]
```

```python
d ={'script':scripts, 'Scheduled languages':languages} 
scripts_df = pd.DataFrame(data=d)
```
