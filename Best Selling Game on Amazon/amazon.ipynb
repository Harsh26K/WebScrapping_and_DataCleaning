```python
URL = 'https://www.amazon.in/gp/bestsellers/videogames/1376528031/ref=zg_bs_pg_1?ie=UTF8&pg=1'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36"}

## request for url
page = requests.get(URL,headers=headers)

## you can use lxml and htmllib5 as well for parsing
soup1 = BeautifulSoup(page.content,'html.parser')

## to make output readable
soup2 = BeautifulSoup(soup1.prettify(),'html')

## we will find all the info of each game at one place
game_info = soup2.find_all('li', class_ = 'zg-item-immersion')

i=0

games = []
reviews= []
total_votes = []
prices=[]
dates = []

## gather the info
for i in range(0,len(game_info)):
   if len(game_info[i].span.div.find_all('a')) == 4:
      game = game_info[i].span.div.find_all('a')[0].find_all('div')[1].get_text().strip()
      review = game_info[i].span.div.find_all('a')[1].span.get_text().strip()
      votes = game_info[i].span.div.find_all('a')[2].get_text().strip()
      price = game_info[i].span.div.find_all('a')[3].span.get_text().strip()
   elif len(game_info[i].span.div.find_all('a')) == 3 :
      game = game_info[i].span.div.find_all('a')[0].find_all('div')[1].get_text().strip()
      review = game_info[i].span.div.find_all('a')[1].span.get_text().strip()
      votes = game_info[i].span.div.find_all('a')[2].get_text().strip() 
      price = 'unavailable'
   elif len(game_info[i].span.div.find_all('a')) == 2: 
      game = game_info[i].span.div.find_all('a')[0].find_all('div')[1].get_text().strip()
      price = game_info[i].span.div.find_all('a')[1].span.get_text().strip()  
      review = 'none'
      votes = 0
   else:
      game = game_info[i].span.div.find_all('a')[0].find_all('div')[1].get_text().strip()
      price = 'unavailable'
      review = 'none'
      votes = 0
   
   today = datetime.date.today()
   # data = [game,price,review,votes,today]
   # with open('BestSellerGames.csv','a+',newline='',encoding='UTF8') as f:
   #     writer = csv.writer(f)
   #     writer.writerow(data)

   print(f'{game} : {review} : {votes} : {price} : {today} \n')

   games.append(game)
   reviews.append(review)
   total_votes.append(votes)
   prices.append(price)
   dates.append(today)

##convert to csv file

d = {'Game':games,'Price':prices,'Review':reviews,'Votes':total_votes,'Date':dates}
df = pd.DataFrame(data=d)
df.to_csv('BestSellerGames.csv')
```
